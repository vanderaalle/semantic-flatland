# Webster 1828 — Definition Fetch

Fetches Webster 1828 definitions from webstersdictionary1828.com for any
corpus passage and writes `{CORPUS_PREFIX}_definitions_w1828.json` to `../data/`.

Requires: `pip install beautifulsoup4`

**Configure** the three lines in the next cell, then run all.

```python
# ── configuration ─────────────────────────────────────────────────────────────
CORPUS_PREFIX = 'potocki'                        # 'potocki' or 'poe'
PASSAGE_FILE  = '../data/corpus/manuscript.txt'  # path to passage text
META_TRIM     = False   # True → strip everything after first blank line

OUT_DIR = '../data'
DELAY   = 1.0   # be polite to the server
```

```python
import sys, os
sys.path.insert(0, '.')

from isotopy import build_lemma_map, fetch_w1828

text = open(PASSAGE_FILE, encoding='utf-8').read().strip()
if META_TRIM:
    text = text.split('\n\n')[0].strip()

lemma_map = build_lemma_map(text)
print(f'{len(lemma_map)} lemmas extracted from passage')
print(sorted(lemma_map)[:10], '...')
```

```python
out = os.path.join(OUT_DIR, f'{CORPUS_PREFIX}_definitions_w1828.json')
result = fetch_w1828(lemma_map=lemma_map, out=out, delay=DELAY)
```

```python
import json
d = json.load(open(out))
empty = [l for l, v in d['definitions'].items() if not v['definition_words']]
print(f"Lemmas: {len(d['definitions'])}   Empty: {empty or 'none'}")

# spot-check a few entries
for lemma in list(d['definitions'])[:5]:
    raws = d['definitions'][lemma]['raw_definitions']
    print(f'  {lemma}: {raws[0][:80] if raws else "EMPTY"}')
```
