# Merriam-Webster — Definition Fetch

Fetches MW Collegiate definitions for any corpus passage and writes
`{CORPUS_PREFIX}_definitions_mw.json` to `../data/`.

**Configure** the three lines in the next cell, then run all.

```python
# ── configuration ─────────────────────────────────────────────────────────────
CORPUS_PREFIX = 'potocki'                        # 'potocki' or 'poe'
PASSAGE_FILE  = '../data/corpus/manuscript.txt'  # path to passage text
META_TRIM     = False   # True → strip everything after first blank line

API_KEY  = 'e403caca-77ca-47cc-a577-c9a2f198e1e2'
OUT_DIR  = '../data'
DELAY    = 0.3
```

```python
import sys, os
sys.path.insert(0, '.')

from isotopy import build_lemma_map, fetch_mw

# load and optionally trim passage
text = open(PASSAGE_FILE, encoding='utf-8').read().strip()
if META_TRIM:
    text = text.split('\n\n')[0].strip()

lemma_map = build_lemma_map(text)
print(f'{len(lemma_map)} lemmas extracted from passage')
print(sorted(lemma_map)[:10], '...')
```

```python
out = os.path.join(OUT_DIR, f'{CORPUS_PREFIX}_definitions_mw.json')
result = fetch_mw(lemma_map=lemma_map, out=out, api_key=API_KEY, delay=DELAY)
```

```python
import json
d = json.load(open(out))
empty = [l for l, v in d['definitions'].items() if not v['definition_words']]
print(f"Lemmas: {len(d['definitions'])}   Empty: {empty or 'none'}")
```
