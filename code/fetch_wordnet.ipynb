# WordNet 3.1 — Definition Fetch

Reads WordNet 3.1 dict files from the local folder and writes
`{CORPUS_PREFIX}_definitions_wn.json` to `../data/`.

**Configure** the three lines in the next cell, then run all.

```python
# ── configuration ─────────────────────────────────────────────────────────────
CORPUS_PREFIX = 'potocki'                          # 'potocki' or 'poe'
PASSAGE_FILE  = '../data/corpus/manuscript.txt'    # path to passage text
META_TRIM     = False   # True → strip everything after first blank line

DICT_DIR = '../data/wn3.1.dict/dict'
OUT_DIR  = '../data'
```

```python
import sys, os
sys.path.insert(0, '.')

from isotopy import build_lemma_map, fetch_wordnet

text = open(PASSAGE_FILE, encoding='utf-8').read().strip()
if META_TRIM:
    text = text.split('\n\n')[0].strip()

lemma_map = build_lemma_map(text)
print(f'{len(lemma_map)} lemmas extracted from passage')
print(sorted(lemma_map)[:10], '...')
```

```python
out = os.path.join(OUT_DIR, f'{CORPUS_PREFIX}_definitions_wn.json')
result = fetch_wordnet(lemma_map=lemma_map, dict_dir=DICT_DIR, out=out)
```

```python
import json
d = json.load(open(out))
empty = [l for l, v in d['definitions'].items() if not v['definition_words']]
print(f"Lemmas: {len(d['definitions'])}   Empty/missing: {empty or 'none'}")
```
